# 01 Data loading

This notebook documents the planned data loading procedure for the Python-based fNIRS analysis pipeline.

The goal of this step is to describe how local fNIRS data will be loaded, how the basic structure of each file will be inspected, and how the data will be prepared for later preprocessing and GLM analysis.


## Data source

The data used in this project come from typically developing children who completed language-related fNIRS tasks.

Participants will be divided into two reading-development groups:

- Lower-grade group: Grades 1-3
- Upper-grade group: Grades 4-6

The main purpose of the project is to build a Python-based fNIRS analysis pipeline and use it to compare brain activation patterns between these two developmental stages.


## Data privacy and availability

The original fNIRS data are not uploaded to this public GitHub repository because the dataset contains data from child participants and may include sensitive information.

This repository only includes code, documentation, example folder structure, and non-sensitive outputs.

To reproduce the analysis, users need to place their own local fNIRS data in the `data/` folder.


## Expected local data structure

The expected local data structure is:

```text
data/
├── raw/
│   ├── sub-001/
│   ├── sub-002/
│   └── ...
├── processed/
└── derivatives/
```

The `data/raw/` folder is ignored by Git and should only exist on the local computer.


## Planned data loading steps

The planned data loading procedure includes:

1. Define the local data directory.
2. Read fNIRS files using MNE-Python or MNE-NIRS compatible functions.
3. Inspect channel names and channel types.
4. Check sampling rate.
5. Check stimulus annotations or event markers.
6. Confirm participant grouping information.
7. Prepare the raw data object for preprocessing.
