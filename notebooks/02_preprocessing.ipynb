02 Preprocessing

This notebook documents the planned preprocessing procedure for the fNIRS analysis pipeline.

The goal of this step is to transform raw fNIRS signals into data suitable for subject-level GLM analysis.


Preprocessing goals

The preprocessing stage will prepare the raw fNIRS data for later statistical analysis.

The planned preprocessing steps include:

Check stimulus markers.
Identify incomplete or unusable files.
Inspect data quality.
Convert raw intensity signals to optical density.
Convert optical density to hemoglobin concentration.
Prepare HbO and HbR data for GLM analysis.


Relation to the original MATLAB workflow

The current Python pipeline is developed by referring to an existing MATLAB fNIRS workflow.

The MATLAB workflow includes marker checking, short-separation channel labeling, resampling, optical density conversion, Beer-Lambert Law conversion, baseline trimming, and later GLM analysis.

The Python version will aim to reproduce the logic of this workflow while using MNE-Python and MNE-NIRS when possible.


Planned preprocessing outputs

The expected outputs of this notebook include:

preprocessed fNIRS objects
preprocessing logs
data quality summary
saved intermediate files in the local data/processed/ folder
