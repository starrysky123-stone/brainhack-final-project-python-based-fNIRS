# 00 Project overview

This notebook provides an overview of the BrainHack final project.

The project aims to build a Python-based fNIRS analysis pipeline for comparing brain activation patterns between lower-grade and upper-grade children during language-related tasks.


## Project goal

The main goal is to document and gradually implement a reproducible fNIRS analysis workflow.

The planned workflow includes:

1. Data loading
2. fNIRS preprocessing
3. Subject-level GLM
4. Group-level comparison
5. Visualization and output generation


## Research context

The dataset comes from typically developing children who completed language-related fNIRS tasks.

Participants will be divided into two developmental groups:

- Lower-grade group: Grades 1-3
- Upper-grade group: Grades 4-6

The analysis will focus on comparing task-related brain activation between these two groups.


## Open science and data privacy

Because the original data include child participants, raw data and participant information are not uploaded to GitHub.

This repository only includes code, documentation, example project structure, and non-sensitive outputs.


## Planned tools

This project will refer to MNE-Python and MNE-NIRS examples for building the Python-based fNIRS analysis workflow.

The current repository provides a structured starting point. The implementation will be expanded after confirming the original data format and local file organization.
